```python
import sympy as sp

x = sp.symbols('x')
expr = (sp.exp(-2*x) + x) * sp.exp(x)

sp.Integral(expr, (x, 0, 1)).doit()
```

```python
x, y = sp.symbols('x y')
f = x*y

sp.integrate(f, (x, 0, 1), (y, 0, 2))
```

```python
sp.integrate(f, (y, 0, 2), (x, 0, 1))
```

```python
import sympy as sp

x, y = sp.symbols('x y')
expr = y/x

sp.integrate(expr, (y, x, 2*x), (x, 2, 4))
```
