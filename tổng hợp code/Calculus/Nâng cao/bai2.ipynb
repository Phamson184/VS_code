```python
from sympy import Symbol, Derivative

x = Symbol('x')
f = (x**3 + x**2 + x)*(x**2 + x)

Derivative(f, x).doit()
```

```python
import sympy
from sympy import Symbol, Derivative

x = Symbol('x')
f = sympy.sin(2*x)

Derivative(f, x).doit()
```

```python
from sympy import Symbol, Derivative

t = Symbol('t')
st = 5*t**2 + 2*t + 8
Derivative(st, t)
```

```python
d = Derivative(st, t)
d.doit()
```

```python
d.doit().subs({t:1})
```

```python
t1 = Symbol('t1')
d.doit().subs({t:t1})
```

```python
t2 = 10
d.doit().subs({t:t2})
```
