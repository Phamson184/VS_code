```python
import sympy as sp
from sympy import (
    symbols, limit, diff, integrate, solve,
    sin, cos, exp, ln, sqrt, pi, oo
)
import matplotlib.pyplot as plt

sp.init_printing(use_unicode=True)

x, y, t, n = symbols('x y t n', real=True)
```

```python
# a)
lim_a = limit(sin(3*x)/(5*x), x, 0)

# b)
lim_b = limit((3*x**2 - 2*x + 1)/(x**2 + 4*x - 5), x, oo)

# c)
lim_c = limit((x**3 - 1)/(x**2 - 1), x, 1)

lim_a, lim_b, lim_c
```

```python
P = 50_000_000
r = 0.08
t_years = 10

A = P * sp.exp(r * t_years)
A
```

```python
f = (exp(x) - 1)/x
lim_f = limit(f, x, 0)
lim_f
```

```python
# a)
y1 = exp(2*x)/(x**2 + 1)
dy1 = diff(y1, x)

# b)
y2 = ln(sin(x) + cos(x))
dy2 = diff(y2, x)

# c)
y3 = x**x
dy3 = diff(y3, x)

dy1, dy2, dy3
```

```python
f = x**3 - 6*x**2 + 9*x + 1
f1 = diff(f, x)
f2 = diff(f1, x)

critical_points = solve(f1, x)
critical_points
```

```python
# Giá trị hàm tại điểm tới hạn
[(cp, f.subs(x, cp)) for cp in critical_points]
```

```python
# Vẽ đồ thị
xs = sp.linspace(-1, 5, 400)
fs = [f.subs(x, val) for val in xs]

plt.plot(xs, fs)
plt.axhline(0)
plt.axvline(0)
plt.grid()
plt.show()
```

```python
x = symbols('x', positive=True)
V = x * (12 - 2*x)**2

dV = diff(V, x)
opt_x = solve(dV, x)
opt_x
```

```python
V_max = V.subs(x, opt_x[0])
V_max
```

```python
# a)
I1 = integrate(3*x**2 - 2*x + 5, x)

# b)
I2 = integrate(sin(2*x), (x, 0, pi/2))

# c)
I3 = integrate(ln(x)/x, (x, 1, sp.E))

I1, I2, I3
```

```python
y1 = x**2
y2 = x + 2

intersections = solve(y1 - y2, x)
intersections
```

```python
area = integrate(y2 - y1, (x, intersections[0], intersections[1]))
area
```

```python
# Vẽ hình
xs = sp.linspace(-2, 3, 400)
plt.plot(xs, [y1.subs(x,i) for i in xs], label='y=x^2')
plt.plot(xs, [y2.subs(x,i) for i in xs], label='y=x+2')
plt.legend()
plt.grid()
plt.show()
```

```python
v = 3*t**2 - 6*t + 2

# a) Quãng đường
s = integrate(v, (t, 0, 3))

# b) Gia tốc
a = diff(v, t)
a_2 = a.subs(t, 2)

s, a_2
```

```python
x, y = symbols('x y')
z = x**3 + y**3 - 3*x*y

zx = diff(z, x)
zy = diff(z, y)

zxx = diff(zx, x)
zyy = diff(zy, y)
zxy = diff(zx, y)
zyx = diff(zy, x)

zx, zy, zxx, zyy, zxy, zyx
```

```python
expr = x**2 + y**2

I_xy = integrate(integrate(expr, (x, 0, 2)), (y, 0, 3))
I_yx = integrate(integrate(expr, (y, 0, 3)), (x, 0, 2))

I_xy, I_yx
```
