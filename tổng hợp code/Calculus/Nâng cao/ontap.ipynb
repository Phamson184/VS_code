```python
def bai4(n):
    a = 4
    for _ in range(n-1):
        a = 5/(6-a)
    return a

[bai4(i) for i in [5,6,7,10,100]]
```

Bước 1: Tìm giới hạn L (nếu tồn tại)
Giả sử lim(n→∞) a_n = L
→ a_{n+1} → L khi n → ∞
→ L = 5/(6 - L)
→ L(6 - L) = 5
→ 6L - L² = 5
→ L² - 6L + 5 = 0
→ (L - 1)(L - 5) = 0
→ L = 1 hoặc L = 5
Bước 2: Xác định L nào đúng
Kiểm tra:
a_1 = 4
a_2 = 5/(6-4) = 5/2 = 2.5
a_3 = 5/(6-2.5) = 5/3.5 ≈ 1.43
# → Dãy giảm và > 1 → L = 1 ✓

# Nếu a_1 = 5:
a_2 = 5/(6-5) = 5/1 = 5
# → Dãy không đổi → L = 5
Bước 3: Chứng minh hội tụ bằng quy nạp
Chứng minh: 1 < a_n < 5 với mọi n ≥ 2

Cơ sở: a_2 = 2.5, 1 < 2.5 < 5 ✓

Quy nạp: Giả sử 1 < a_n < 5
→ 6 - 5 < 6 - a_n < 6 - 1
→ 1 < 6 - a_n < 5
→ 5/5 < 5/(6 - a_n) < 5/1
→ 1 < a_{n+1} < 5 ✓

→ Dãy bị chặn

Chứng minh dãy đơn điệu:
a_{n+1} - a_n = 5/(6 - a_n) - a_n
             = [5 - a_n(6 - a_n)]/(6 - a_n)
             = [5 - 6a_n + a_n²]/(6 - a_n)
             = (a_n² - 6a_n + 5)/(6 - a_n)
             = (a_n - 1)(a_n - 5)/(6 - a_n)

Với 1 < a_n < 5:
- (a_n - 1) > 0
- (a_n - 5) < 0
- (6 - a_n) > 0
→ a_{n+1} - a_n < 0
→ Dãy giảm

Dãy giảm + bị chặn → hội tụ → L = 1 ✓
