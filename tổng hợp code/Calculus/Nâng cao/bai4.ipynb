```python
import matplotlib.pyplot as plt

months = list(range(1,13))
rain_days = [3,1,4,9,18,19,21,21,20,19,13,7]

plt.figure()
plt.plot(months, rain_days, marker='o')
for i, txt in enumerate(rain_days):
    plt.text(months[i], rain_days[i]-1.2, str(txt), ha='center')

plt.xticks(months)
plt.xlabel("Tháng")
plt.ylabel("Số ngày mưa")
plt.title("Số ngày mưa trung bình theo tháng - TP.HCM")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

months = list(range(1,13))
sun_hours = [8.8,9.3,9.2,9.2,9.0,9.0,8.9,8.6,8.1,7.9,8.2,8.2]

plt.figure()
plt.plot(months, sun_hours, marker='o')

for i, txt in enumerate(sun_hours):
    if sun_hours[i] < 8.1:
        plt.text(months[i], sun_hours[i]+0.15, str(txt), ha='center')
    else:
        plt.text(months[i], sun_hours[i]-0.25, str(txt), ha='center')
        
plt.xticks(months)
plt.xlabel("Tháng")
plt.ylabel("Giờ nắng trung bình")
plt.title("Số giờ nắng trung bình theo tháng - TP.HCM")

plt.grid(True)
plt.show()
```
