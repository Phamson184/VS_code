```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 400)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.sin(2*x), label='sin(2x)')
plt.plot(x, np.sin(3*x), label='sin(3x)')

plt.legend()
plt.show()
plt.savefig("basics_python.pdf")
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
class SinPlot:
    def __init__(self,freq):
        self.freq=freq
    def draw(self):
        x=np.linspace(0,2*np.pi,400) 
        y=np.sin(self.freq*x) 
        plt.plot(x,y,label=f"sin({self.freq}x)")  
p=SinPlot(2)
p.draw()
plt.legend()
plt.show()
plt.close()        
```

```python
import math
def traprule(ham, trai, phai):
    return (phai - trai) * (ham(trai) + ham(phai)) / 2

print("Bieu thuc can tinh tich phan la e^x [a, b]")

a = float(input("Nhap a : "))
b = float(input("Nhap b : "))

xap_xi = traprule(math.exp, a, b)

chinh_xac = math.exp(b) - math.exp(a)

print(f"Bieu thuc can tinh tich phan la e^x [{a:.1E},{b:.1E}] :")
print(f"Gia tri xap xi    : {xap_xi:.15E}")
print(f"Gia tri chinh xac : {chinh_xac:.15E}")
```
